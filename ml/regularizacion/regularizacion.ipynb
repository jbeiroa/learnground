```python
from sklearn.datasets import load_diabetes
import pandas as pd

# Cargar el conjunto de datos de diabetes
data = load_diabetes()

# Convertir a DataFrame para una mejor visualizaci√≥n
df_diabetes = pd.DataFrame(data.data, columns=data.feature_names)
df_diabetes['target'] = data.target

# Mostrar las primeras filas del conjunto de datos
df_diabetes.head()
```

### Ejercicio 1 - Regularizaci√≥n utilizando Ridge

Aplica regularizaci√≥n Ridge (L2) en un modelo de regresi√≥n lineal para predecir la columna 'target' usando todas las caracter√≠sticas del conjunto de datos. Divide los datos en conjuntos de entrenamiento y prueba. Imprime el error cuadr√°tico medio de entrenamiento y de prueba. Discute los resultados al variar el par√°metro alpha.
  

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X = df_diabetes.drop('target', axis=1)
y = df_diabetes['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Lista de alphas a probar
n_alphas = 10
alphas = np.logspace(-5,0, n_alphas)

coefs = []
for a in alphas:
    ridge = Ridge(alpha=a, fit_intercept=False)
    ridge.fit(X_train, y_train)
    coefs.append(ridge.coef_)
```

```python
coeficientes_finales = pd.DataFrame([np.array(X_train.columns.tolist()), coefs]).T
coeficientes_finales.columns = ['feature','coeficiente']
coeficientes_finales
```

```python
coefs[0], alphas[0]
```

```python
#MSE para el conjunto de entrenamiento
preds_train = []
for a in alphas:
    ridge = Ridge(alpha=a, fit_intercept=False)
    ridge.fit(X_train, y_train)
    y_ = ridge.predict(X_train)
    preds_train.append(y_)
```

```python
mses_train = [mean_squared_error(y_train, pred) for pred in preds_train]
mses_train
```

```python
#MSE para el conjunto de prueba
preds_test = []
for a in alphas:
    ridge = Ridge(alpha=a, fit_intercept=False)
    ridge.fit(X_train, y_train)
    y_ = ridge.predict(X_test)
    preds_test.append(y_)

mses_test = [mean_squared_error(y_test, pred) for pred in preds_test]
mses_test
```

## Ejercicio 2 - Validaci√≥n cruzada (Cross Validation)
   
Realizar una validaci√≥n cruzada utilizando K-Fold con K = 10 para el modelo de regresi√≥n lineal con el conjunto de datos "Diabetes". Discutir los resultados.

  

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

# Inicializar el modelo de regresi√≥n lineal
lm = LinearRegression()
# Realizar validaci√≥n cruzada de 10-folds
scores = cross_val_score(lm, X_train, y_train, cv=10, scoring='neg_mean_squared_error')
# Calcular la media y desviaci√≥n est√°ndar de MSE
print(f'La media del MSE es {np.negative(scores).mean()}.')
print(f'La std del MSE es {np.negative(scores).std()}.')
```

## Ejercicio 3 - Afinaci√≥n de hiperpar√°metros con RandomizedSearch
   
Realizar una b√∫squeda de la mejor combinaci√≥n de hiperpar√°metros para un modelo de regresi√≥n basado en √°rboles utilizando RandomizedSearchCV con el conjunto de datos "Diabetes". Discutir los resultados.

  

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV

# Par√°metros para RandomizedSearchCV
parametros = {'n_estimators': np.arange(50, 100, 1),
              'min_samples_leaf': np.arange(1, 10, 1),
              'bootstrap': [True, False],
              'max_depth': np.arange(10, 100, 10)}
# Inicializar RandomForestRegressor
rf = RandomForestRegressor()
# Inicializar RandomizedSearchCV
rf_random = RandomizedSearchCV(rf, param_distributions=parametros, n_iter=50, scoring='neg_mean_squared_error')
# Ajustar el modelo
rf_random.fit(X_train, y_train)
# Extraer los mejores par√°metros y puntuaci√≥n
cv_results = rf_random.cv_results_
for mean_score, params in zip(cv_results["mean_test_score"], cv_results["params"]):
    print(np.sqrt(-mean_score), params)
```
