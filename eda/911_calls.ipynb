## Creating new features
In the titles column there are "Reasons/Departments" specified before the title code. These are EMS, Fire, and Traffic. Let's create a new column called _Reason_ that contains this string value by using .apply() with a custom lambda expression.

__For example, if the title column value is EMS: BACK PAINS/INJURY , the Reason column value would be EMS.__

```python
df['Reason'] = df['title'].apply(lambda title: title.split(":")[0])
df['Reason']
```

__What is the most common Reason for a 911 call based off of this new column?__

```python
df['Reason'].value_counts()
```

__Create a countplot of 911 calls by Reason.__

```python
sns.countplot(x='Reason', data=df, palette='viridis')
```

__Now let us begin to focus on time information.__

The data type of the objects in the timeStamp column is str.

```python
type(df['timeStamp'].iloc[0])
```

Let's convert the column from str to DateTime objects with pd.to_datetime.

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

Now, let's create 3 new columns called Hour, Month, and Day of Week.

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['Hour']=df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time: time.day_of_week).map(dmap)
```

__Use seaborn to create a countplot of the Day of Week column with the hue based off of the Reason column.__

```python
sns.countplot(x='Day of Week', 
              data=df, 
              hue='Reason', 
              palette='viridis',
              order=dmap.values())
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

__Now the same for month.__

```python
sns.countplot(x='Month', data=df, hue='Reason', palette='viridis')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

It seems there's a few months missing (9, 10 and 11). Let's see a few ways to impute this values.

Graphically, a simple lineplot can be used to "more or less" fill the missing points. Let's group the data by the month value and use the count( aggregate function. 

```python
byMonth = df.groupby('Month').count()
byMonth
```

__Now we can plot any column to get a feeling for the missing values.__

```python
# Could be any column
byMonth['twp'].plot()
```

__We can also create a linear fit on the number of calls per month with seaborn's lmplot(). The data is clearly non-linear, though.__

```python
plot = sns.lmplot(x='Month',y='twp', data=byMonth.reset_index())
plot.ax.set_xticks(np.arange(1, 13, 1));
```

## More data visualization

__Create a new column called 'Date' that contains the date from the timeStamp column.__

```python
df['Date']=df['timeStamp'].apply(lambda t: t.date())
```

__Now groupby this Date column with the count() aggregate and create a plot of counts of 911 calls.__

```python
df.groupby('Date').count()['twp'].plot()
plt.tight_layout()
```

__Now let's plot one line for each Reason.__

```python
for reason in df['Reason'].unique():
    df[df['Reason']==reason].groupby('Date').count()['twp'].plot(label=reason)
plt.title('911 Calls by Date')
plt.legend()
plt.tight_layout()
```

__Now let's move on to creating heatmaps with seaborn and our data.__

Let's begin by creating a heatmp to show at which hours are made most of the calls everyday.
We'll first need to restructure the dataframe so that the columns become the Hours and the Index becomes the Day of the Week.

```python
dayHour = df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayHour,cmap='viridis')
```

```python
sns.clustermap(dayHour,cmap='viridis')
```

At last, let's do the same but grouping by month.

```python
dayMonth = df.groupby(by=['Day of Week','Month']).count()['Reason'].unstack()
dayMonth
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayMonth,cmap='viridis')
```

```python
sns.clustermap(dayMonth,cmap='viridis')
```
